```python
import pandas as pd

# load the file stereotypes_global_v2.csv

df = pd.read_csv('stereotypes_global_v2.csv', encoding='latin-1', sep=';')
display(df)
```

```python
# we filter the dataset to considere british stereotypes
df = df[df['identity'] == 'British']
display(df)

# we sort on the region_stereo column
df = df.sort_values(by=['region_stereo'], ascending=False)
display(df.head(10))

# we sort on the NA_stereo column
df = df.sort_values(by=['NA_stereo'], ascending=False)
display(df.head(10))
```
